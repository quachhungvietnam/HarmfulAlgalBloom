```python
def Clean_date(df):
    for index, value in enumerate(df["Collection_Date"]):
        df["Collection_Date"][index] = datetime.strptime(value, '%Y-%m-%d %H:%M:%S')        
    return df
```

```python
for station in Site_name:
    data = pd.read_csv(station+".csv")
    #Transfer text in "Collection_Date" to datetime value:
    data = Clean_date(data)

    if station in latlondict:
        lat, lon = latlondict[station]
    else:
        lat, lon = None, None
        Future_data.append(station)

    Lat.append(lat)
    Lon.append(lon)
    
    if len(data) > 1:
        Begin_date.append(min(data["Collection_Date"]))
        End_date.append(max(data["Collection_Date"]))
    else:
        Begin_date.append(None)
        End_date.append(None)
        
    Total_measurements.append(len(data))
```

```python
#Column: Station
Station_array = []
#Column: Variable
Variable_array = []
#Column: Begin date
Begin_array = []
#Column: End date
End_array = []
#Column: Total measurements
Total_measurements_array = []
#Column: Measurments since 2000
Measurements_since_2000_array = []
#Column: Measurements in 2019:
Measurements_in_2019_array = []

for station in Station_list:
    #Request data from local folder:
    data = pd.read_csv(station+".csv")
    #Clean date data:
    data = Clean_date(data)
    
    for variable in Variable_list:
        data_for_variable = data.loc[data["Test Name"]==variable,:]
        
        #Add value to each column of the table:
        Station_array.append(station)
        
        Variable_array.append(variable)
        if len(data_for_variable) > 2:
            Begin_array.append(min(data_for_variable["Collection_Date"]))
            End_array.append(max(data_for_variable["Collection_Date"]))
        else:
            Begin_array.append(None)
            End_array.append(None)
            
        Total_measurements_array.append(len(data_for_variable))
        
        index_since_2000 = [True if x.year >= 2000 else False for x in data_for_variable["Collection_Date"]]
        Measurements_since_2000_array.append(len(data_for_variable.loc[index_since_2000,:]))
        
        index_in_2019 = [True if x.year == 2019 else False for x in data_for_variable["Collection_Date"]]
        Measurements_in_2019_array.append(len(data_for_variable.loc[index_in_2019,:]))
```

```python
Summary_table = pd.DataFrame({"Station": Station_array,
                             "Variable": Variable_array,
                             "Begin date": Begin_array,
                             "End date": End_array,
                             "Total measurements": Total_measurments_array,
                             "Measurements since 2000": Measurements_since_2000_array,
                             "Measurements in 2019": Measurments_in_2019_array})
```
