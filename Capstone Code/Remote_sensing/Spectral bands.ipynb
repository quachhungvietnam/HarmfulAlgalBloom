```python
#Sentinel 2:
central_width = [[492.4,66], [559.8,36], [664.6, 31], [832.8, 106],
                [704.1, 15], [740.5, 15], [782.8, 20], [864.7, 21],
                [442.7,21], [945.1,20], [1373.5,31]]
spectral["SENTINEL-2"] = []
for [center, width] in central_width:
    spectral["SENTINEL-2"].append([center-width/2,center+width/2])
```

```python
for satellite in spectral.keys():
    fig, ax = plt.subplots(figsize=(15,1))
    ax.plot([400],[0])
    ax.plot([1000],[1])
    
    for spectral_range in spectral[satellite]:
        x, y = spectral_range[0], 0
        width = spectral_range[1] - spectral_range[0]
        ax.add_patch(
             patches.Rectangle(
                (x, y),
                width,
                10,
                edgecolor = 'white',
                facecolor = 'blue',
                fill=True      
             ) )
        
    ax.set_xlabel("Wavelength (nm)")
    ax.set_yticks([])
    ax.set_xticks(np.arange(400,1001,50))
    print(satellite)
    #ax.set_title(satellite)

    plt.show()
```
